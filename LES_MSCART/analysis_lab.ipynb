```python
#fig13 ,fig13_ttl =DYCOMS2_3D_1D_bias(DYC0p860_140,0,140,6)
case=DYC0p860_140;VZAi=61
vc=case.get_bin_defs(6)
RTbias=POLCARTdset('LES','')
RTbias.MeanPRad=case.RT.MeanPRad-case.RT1D.MeanPRad

fig1,ax1=plt.subplots(2,3,figsize=(12,7),subplot_kw={'aspect':'equal'})
fig1_ttl=case.RT.fname.split('.',1)[0]+'_DYCOM2_IQU_3D_1D_biases'
#fig1.suptitle("\n".join(wrap(fig1_ttl,50)),size=10)

ctfI=ax1[0,0].contourf(case.xcens,case.ycens,case.RT.MeanPRad[VZAi,:,:,0]  ,vc['vIR'],cmap=plt.cm.jet,extend='both')
ctI1=ax1[0,1].contourf(case.xcens,case.ycens,(case.RT1D.MeanPRad[VZAi,:,:,0]),vc['vIR'],cmap=plt.cm.jet,extend='both')
ctfQ=ax1[1,0].contourf(case.xcens,case.ycens,case.RT.MeanPRad[VZAi,:,:,1]  ,vc['vQR'],cmap=plt.cm.jet,extend='both')
ctQ1=ax1[1,1].contourf(case.xcens,case.ycens,(case.RT1D.MeanPRad[VZAi,:,:,1]),vc['vQR'],cmap=plt.cm.jet,extend='both')
ax1[0,0].set_title('I_3D')
ax1[0,1].set_title('I_1D')
ax1[1,0].set_title('Q_3D')
ax1[1,1].set_title('Q_1D')

ctfb1=ax1[0,2].contourf(case.xcens,case.ycens,RTbias.MeanPRad[VZAi,:,:,0],np.linspace(-.1,.1,50)  ,cmap=plt.cm.RdBu_r,extend='both')
ctfb2=ax1[1,2].contourf(case.xcens,case.ycens,RTbias.MeanPRad[VZAi,:,:,1],np.linspace(-.01,.01,50),cmap=plt.cm.RdBu_r,extend='both')
ax1[0,2].set_title('I Bias')
ax1[1,2].set_title('Q Bias')

ax1[0,0].set_ylabel('km',size=10)
ax1[1,0].set_ylabel('km',size=10)
ax1[1,0].set_xlabel('km',size=10)
ax1[1,1].set_xlabel('km',size=10)
ax1[1,2].set_xlabel('km',size=10)

ax1[0,1].tick_params(labelleft=False)
ax1[0,2].tick_params(labelleft=False)
ax1[1,1].tick_params(labelleft=False)
ax1[1,2].tick_params(labelleft=False)

cpn.add_cb(fig1,ctfI,ax1[0,0],ticks=vc['cIR'],label='',orientation='vertical')
cpn.add_cb(fig1,ctfQ,ax1[1,0],ticks=vc['cQR'],label='',orientation='vertical')
cpn.add_cb(fig1,ctI1,ax1[0,1],ticks=vc['cIR'],label='',orientation='vertical')
cpn.add_cb(fig1,ctQ1,ax1[1,1],ticks=vc['cQR'],label='',orientation='vertical')
cpn.add_cb(fig1,ctfb1,ax1[0,2],ticks=np.arange(-.1,.11,.1) ,label='',orientation='vertical')
cpn.add_cb(fig1,ctfb2,ax1[1,2],ticks=np.arange(-.01,.011,.01),label='',orientation='vertical')
cpn.sub_labels(ax1,x=0.01,y=0.85)
fig1.tight_layout()
```
